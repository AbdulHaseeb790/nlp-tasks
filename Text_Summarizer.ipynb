```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
```

```python
text='''
Pakistan is a country in South Asia. It is the fifth-most populous country
with a population of over 220 million. It is the 33rd-largest country by area.
Pakistan has a long history and rich culture. The country gained independence
in 1947. It has four provinces and several territories. Karachi is the largest
city and the economic hub of the country.
'''
```

```python
print(summarizer(text,max_length=50, min_length=20))
```
